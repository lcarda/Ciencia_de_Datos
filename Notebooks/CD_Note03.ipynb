**b)** el estimador de maxima verosimilitud $\hat{\theta} $ cumple que maximiza el producto de las $p(x_i|\theta)$con $i = 1, ..., n$, y si estos no son ceros se da que:
$$\prod_{i=1}^{n} p(x_i|\theta) = \theta^n e^{-\theta\sum_{i=1}^{n} x_i} $$
derivando el segundo termino
$$ (n\theta^{n-1} - \theta^n \sum_{i=1}^{n} x_i) e^{-\theta\sum_{i=1}^{n} x_i} $$
igualando a 0 (para obtener un punto critico), puedo olvidarme del $e^{-\theta\sum_{i=1}^{n} x_i}$, lo cual me deja con
$$ 0 = n\theta^{n-1} - \theta^n \sum_{i=1}^{n} x_i $$
$$ 0 = n - \theta \sum_{i=1}^{n} x_i $$
y despejando tengo
$$ \theta=\frac{1}{\frac{1}{n}\sum_{k=1}^{n}{x_k}} $$
faltaria ver que el tal $\hat \theta$ es maximo pero eso se ve del grafico de $\theta$ del (a) que es parecida a una concava.

Para estar seguro de esto, necesito que $\theta \ge max $ $ x_i$

Ahora, si $\hat \theta = max $ $ x_i$ y sea $\theta > \hat \theta$
$$ p(x_i|\theta) = \frac{1}{\theta} < \frac{1}{\hat\theta} = p(x_i|\hat\theta) $$
esto se daria para toda $i = 1, \dots, n$. Por lo tanto, $\hat \theta$ es maximal.

```python
## 2D discriminant functions

w1dt=train1
w2dt=train2

mu1 = np.array(w1dt[['x1','x2','x3']].mean())
mu2 = np.array(w2dt[['x1','x2','x3']].mean())

Sigma1 = np.array(w1dt[['x1','x2','x3']].cov()*(n-1)/n) # Multiplicación de matrices, da sigma al cuadrado
Sigma2 = np.array(w2dt[['x1','x2','x3']].cov()*(n-1)/n) # Multiplicación de matrices, da sigma al cuadrado

W1 = -1/2*(np.linalg.inv(Sigma1))
W2 = -1/2*(np.linalg.inv(Sigma2))

w1 = np.dot(np.linalg.inv(Sigma1),mu1) #Ec 61 pag 39 Duda
w2 = np.dot(np.linalg.inv(Sigma2),mu2) #Ec 61 pag 39 Duda

w1t = np.transpose(w1)
w2t = np.transpose(w2)

pw1=n1/n
pw2=n2/n


w10 = - np.dot(np.dot(np.transpose(mu1),np.linalg.inv(Sigma1)),mu1) + np.log(np.linalg.det(Sigma1))/2 + np.log(pw1)
w20 = - np.dot(np.dot(np.transpose(mu2),np.linalg.inv(Sigma2)),mu2) + np.log(np.linalg.det(Sigma2))/2 + np.log(pw2)


def g1(X) :
  return  np.dot(X ,np.dot(W1 ,X)) + np.dot(w1t,X) + w10
#
def g2(X) :
  return  np.dot(X ,np.dot(W2 ,X)) + np.dot(w2t,X) + w20
```

```python
#  dicotomizador gaussiano
#
def classifier2D(X) :
  if  g1(X) > g2(X) :
      return 1
  else :
      return 2
```

```python
## Classification 2D
#
ntest=int(ntest/2)
Class1 = list(map(lambda j: classifier2D(np.array(test1.iloc[j,:])), list(range(0, ntest))))
Class2 = list(map(lambda j: classifier2D(np.array(test2.iloc[j,:])), list(range(0, ntest))))


Class1 = np.array(Class1)
Class2 = np.array(Class2)

y_pred=np.append(Class1,Class2)

y_test=np.append(np.ones(np.shape(Class1)),2*np.ones(np.shape(Class1)))
print(Class1,Class2)
#
error2D = np.concatenate((Class1 != 1, Class2 != 2))

nerror2D = len(error2D[error2D==True])
print(nerror2D)
#
ferror2D = nerror2D/ (ntest)
print(ferror2D)
```

```python
#CASO 3 DEL LIBRO APRENDE
def modelo_bay(w1dt,w2dt,n,n1,n2):
  #come los train1 y train2 n,n1,n2

  mu1 = np.array(w1dt.mean())
  mu2 = np.array(w2dt.mean())

  Sigma1 = np.array(w1dt.cov()*(n-1)/n) # Multiplicación de matrices, da sigma al cuadrado, esta formula me da la muestral y yo quiero la muestral
  Sigma2 = np.array(w2dt.cov()*(n-1)/n) # Multiplicación de matrices, da sigma al cuadrado
#Yo quiero la muestral y verosimilitud me da la poblacional.
  W1 = -1/2*(np.linalg.inv(Sigma1))
  W2 = -1/2*(np.linalg.inv(Sigma2))

  w1 = np.dot(np.linalg.inv(Sigma1),mu1) #Ec 68 pag 41
  w2 = np.dot(np.linalg.inv(Sigma2),mu2)

  w1t = np.transpose(w1)
  w2t = np.transpose(w2)

  pw1=n1/n
  pw2=n2/n

  w10 = - (np.dot(np.dot(np.transpose(mu1),np.linalg.inv(Sigma1)),mu1) + np.log(np.linalg.det(Sigma1)))/2 + np.log(pw1)
  w20 = - (np.dot(np.dot(np.transpose(mu2),np.linalg.inv(Sigma2)),mu2) + np.log(np.linalg.det(Sigma2)))/2 + np.log(pw2)
  return (W1,w1t,w10,W2,w2t,w20)
```
