SVC es una implementación de SVM que admite tanto clasificación binaria como multiclase.

**-**Los kernels disponibles en SVC son: lineal, polinomial, radial basis function (RBF), sigmoide y kernels personalizados definidos por el usuario.

**Pros:** Puede manejar conjuntos de datos no linealmente separables utilizando kernels no lineales. Tiene una mayor flexibilidad en la elección de los kernels.

**Contras:** Puede ser más lento en conjuntos de datos grandes. Requiere ajustar los hiperparámetros del kernel y la penalización.

**Parametros:**

*   **C:** Parámetro de regularización. Controla la compensación entre maximizar el margen y minimizar la clasificación incorrecta.
*   **kernel:** Tipo de kernel a utilizar, como 'linear', 'rbf'(defaut), 'poly' o 'sigmoid'.

La función de clasificación de vectores de soporte (SVC) en la biblioteca `scikit-learn` de Python admite varios tipos de kernel para realizar la transformación no lineal de los datos de entrada en un espacio de mayor dimensionalidad. A continuación, te explico brevemente cada uno de los kernels admitidos por SVC:

1. **Linear (lineal)**: Este kernel realiza una transformación lineal de los datos de entrada a un espacio de mayor dimensionalidad. Es útil cuando se desea una clasificación lineal en el espacio de características original.

2. **Poly (polinómico)**: Este kernel realiza una transformación polinómica de los datos de entrada a un espacio de mayor dimensionalidad. El grado del polinomio se especifica con el parámetro `degree`. Es útil cuando se sospecha que la relación entre las características y la variable objetivo es polinómica.

3. **RBF (función de base radial)**: Este kernel utiliza una función de base radial (RBF) para realizar una transformación no lineal de los datos de entrada a un espacio de mayor dimensionalidad. La función RBF es una función de similitud que mide la distancia entre dos puntos en función de su similitud. Es el kernel predeterminado en SVC y generalmente se utiliza cuando no se tiene un conocimiento previo sobre la estructura del problema.

4. **Sigmoid (sigmoide)**: Este kernel realiza una transformación utilizando la función sigmoide de los datos de entrada a un espacio de mayor dimensionalidad. Es útil cuando se sospecha que la relación entre las características y la variable objetivo sigue una función sigmoide.

5. **Precomputed (precalculado)**: Este kernel permite que los datos de entrada ya estén precalculados y pasados como una matriz de similitud. En este caso, se debe proporcionar una matriz de similitud precalculada en lugar de los datos de entrada originales.

Además de estos kernels predefinidos, también es posible especificar un kernel personalizado utilizando una función callable. Esta función callable debe aceptar dos matrices de características como argumentos y devolver una matriz de similitud entre las características.

En resumen, los kernels en SVC permiten realizar transformaciones no lineales de los datos de entrada a un espacio de mayor dimensionalidad, lo que facilita la clasificación en problemas más complejos. La elección del kernel depende del conocimiento previo sobre la estructura del problema y la relación esperada entre las características y la variable objetivo.

*   **degree:** Grado del polinomio en el caso del kernel polinomial.
*   **gamma:** Coeficiente del kernel para 'rbf', 'poly' y 'sigmoid'. Controla la influencia de cada ejemplo de entrenamiento.

El parámetro `gamma` en la función de clasificación de vectores de soporte (SVC) de `scikit-learn` es un coeficiente utilizado por los kernels `'rbf'`, `'poly'` y `'sigmoid'`. Controla la influencia de cada ejemplo de entrenamiento en la forma de la frontera de decisión.

En la documentación de `SVC`, se explican las diferentes opciones para especificar el valor de `gamma`:

1. `'scale'` (escala): Si se pasa `'scale'` como valor de `gamma` (valor predeterminado), se calculará automáticamente como `1 / (n_features * X.var())`, donde `n_features` es el número de características (columnas) en los datos de entrenamiento `X` y `X.var()` es la varianza de `X`.

2. `'auto'`: Si se pasa `'auto'` como valor de `gamma`, se calculará automáticamente como `1 / n_features`, es decir, el inverso del número de características.

3. Valor numérico: Si se pasa un valor numérico no negativo como `gamma`, este valor se utilizará directamente.

En general, el valor de `gamma` controla la flexibilidad del modelo y la influencia de cada ejemplo de entrenamiento en la frontera de decisión. Un valor más pequeño de `gamma` da como resultado una frontera de decisión más suave y extendida, lo que puede llevar a un mayor sesgo pero una menor varianza. Un valor más grande de `gamma` da como resultado una frontera de decisión más ajustada y ajustada a los datos de entrenamiento, lo que puede llevar a un mayor ajuste pero a un mayor riesgo de sobreajuste.

La elección del valor de `gamma` depende del problema y los datos específicos. Si no estás seguro, los valores predeterminados `'scale'` o `'auto'` suelen funcionar bien en muchos casos, ya que se adaptan a las características de los datos automáticamente. Sin embargo, experimentar con diferentes valores de `gamma` puede ayudar a encontrar el mejor equilibrio entre sesgo y varianza para un problema dado.


*   **coef0:** Término independiente en las funciones de kernel polinomial y sigmoidal.
*   **shrinking:** Especifica si utilizar la heurística de reducción de margen para acelerar el entrenamiento.
*    **probability:** Especifica si habilitar la estimación de probabilidades de clase.

a) Diferencias entre los modelos SVM en scikit-learn:

Support Vector Classification (SVC): Es el modelo SVM tradicional que permite utilizar diferentes kernels, como el kernel lineal, el kernel polinómico y el kernel de base radial (RBF). Pros: Puede manejar tanto problemas de clasificación lineal como no lineal. Contras: Puede ser computacionalmente costoso para conjuntos de datos grandes.

Linear Support Vector Classification (LinearSVC): Es una implementación más eficiente de SVM para el caso de clasificación lineal. Utiliza una función de pérdida lineal y no admite el uso de kernels. Pros: Es más rápido que SVC para conjuntos de datos grandes con clasificación lineal. Contras: Solo es adecuado para problemas de clasificación lineal.

Nu-Support Vector Classification: Es similar a SVC, pero utiliza un parámetro de tolerancia "nu" en lugar del parámetro de penalización "C". Permite utilizar los mismos kernels que SVC. Pros: Permite controlar el equilibrio entre el número de vectores de soporte y el margen. Contras: Puede ser menos intuitivo que utilizar el parámetro de penalización "C" en SVC.

Linear classifiers with SGDClassifier: Es una implementación de SVM que utiliza el algoritmo de gradiente descendente estocástico (SGD) para entrenar modelos lineales. Puede utilizarse tanto para problemas de clasificación lineal como no lineal utilizando funciones de pérdida específicas. Pros: Eficiente en términos computacionales y escalable para conjuntos de datos grandes. Contras: Requiere una selección cuidadosa de la función de pérdida y los hiperparámetros para obtener buenos resultados.

b) Esquemas de clasificación multiclase:

SVC: Implementa el esquema one-vs-one para la clasificación multiclase. Crea un clasificador binario para cada par de clases y realiza la clasificación mediante votación.

LinearSVC: Implementa el esquema one-vs-rest (one-vs-all) para la clasificación multiclase. Entrena un clasificador binario para cada clase y asigna una instancia a la clase con la puntuación más alta.

Nu-Support Vector Classification y SGDClassifier: No tienen un esquema específico implementado para clasificación multiclase. Se pueden utilizar técnicas de reducción de problemas multiclase a problemas binarios, como one-vs-one o one-vs-rest, utilizando estas implementaciones junto con otras estrategias.

c) Significado de los parámetros:

C: Es el parámetro de penalización utilizado para controlar el equilibrio entre el margen y el error de clasificación en los modelos SVC y LinearSVC. Un valor más alto de C da más importancia a clasificar correctamente cada punto de datos y puede llevar a límites de decisión más complejos.

nu: Es un parámetro utilizado en el modelo Nu-Support Vector Classification que controla la cantidad de vectores de soporte y el margen. Un valor más bajo de nu permite un mayor número de vectores de soporte y un margen más amplio.

gamma: Es el parámetro del kernel utilizado en los modelos SVM con kernels (SVC, Nu-SVC). Controla la influencia de cada ejemplo de entrenamiento en el modelo final. Un valor más alto de gamma da más peso a los ejemplos cercanos y puede llevar a límites de decisión más ajustados.

coef0: Es un parámetro utilizado en los kernels polinómicos y sigmoidales en los modelos SVM. Controla la influencia de los términos de grado más alto en la función de decisión.

degree: Es el grado del kernel polinómico utilizado en los modelos SVM con kernel polinómico. Especifica el grado del polinomio a utilizar en la función de decisión.

class_weight: Es un parámetro que permite asignar pesos diferentes a las clases en función de su frecuencia o importancia relativa. Puede ser útil en problemas con clases desequilibradas para dar más importancia a las clases minoritarias.

Estos parámetros se aplican de manera específica a cada modelo SVM y tienen un impacto en la forma en que se construye el límite de decisión y se realizan las predicciones. Es importante ajustar adecuadamente estos parámetros para obtener un buen rendimiento del modelo en cada problema específico.

````python
import matplotlib.pyplot as plt
from sklearn import svm, datasets
from sklearn.inspection import DecisionBoundaryDisplay


# import some data to play with
iris = datasets.load_iris()
# Take the first two features. We could avoid this by using a two-dim dataset
X = iris.data
y = iris.target
X_2d= X[:, :2]
# we create an instance of SVM and fit out data. We do not scale our
# data since we want to plot the support vectors
C = 0.1  # SVM regularization parameter
models = (
    svm.SVC(kernel="linear", C=C),
    svm.LinearSVC(C=C, max_iter=10000),
    svm.SVC(kernel="rbf", gamma=0.7, C=C),
    svm.SVC(kernel="poly", degree=3, gamma="auto", C=C),
)
models = (clf.fit(X_2d, y) for clf in models)

# title for the plots
titles = (
    "SVC with linear kernel",
    "LinearSVC (linear kernel)",
    "SVC with RBF kernel",
    "SVC with polynomial (degree 3) kernel",
)

# Set-up 2x2 grid for plotting.
fig, sub = plt.subplots(2, 2)
plt.subplots_adjust(wspace=0.4, hspace=0.4)

X0, X1 = X_2d[:, 0], X_2d[:, 1]

for clf, title, ax in zip(models, titles, sub.flatten()):
    disp = DecisionBoundaryDisplay.from_estimator(
        clf,
        X_2d,
        response_method="predict",
        cmap=plt.cm.coolwarm,
        alpha=0.8,
        ax=ax,
        xlabel=iris.feature_names[0],
        ylabel=iris.feature_names[1],
    )
    ax.scatter(X0, X1, c=y, cmap=plt.cm.coolwarm, s=20, edgecolors="k")
    ax.set_xticks(())
    ax.set_yticks(())
    ax.set_title(title)

plt.show()
````

La curva ROC (Receiver Operating Characteristic) es una herramienta utilizada en la evaluación de modelos de clasificación binaria para analizar y visualizar su rendimiento. La curva ROC muestra la relación entre la tasa de verdaderos positivos (TPR) y la tasa de falsos positivos (FPR) a medida que se varía el umbral de clasificación del modelo.

Aquí se explica cómo funciona la construcción de una curva ROC:

1. Predecir probabilidades: En primer lugar, el modelo de clasificación binaria debe generar probabilidades de predicción para cada ejemplo en el conjunto de datos de prueba. Estas probabilidades representan la confianza del modelo en la predicción de cada clase.

2. Establecer umbral de clasificación: Luego, se elige un umbral de clasificación para convertir las probabilidades en etiquetas de clase. Por ejemplo, si el umbral es 0.5, se considera que todas las probabilidades superiores a 0.5 pertenecen a la clase positiva, y las probabilidades inferiores se asignan a la clase negativa.

3. Calcular TPR y FPR: A medida que se varía el umbral de clasificación, se calculan la tasa de verdaderos positivos (TPR) y la tasa de falsos positivos (FPR). El TPR representa la proporción de ejemplos positivos correctamente clasificados como positivos, mientras que el FPR representa la proporción de ejemplos negativos incorrectamente clasificados como positivos.

4. Construir la curva ROC: A medida que se ajusta el umbral de clasificación, se registran los valores de TPR y FPR para diferentes puntos a lo largo de la curva ROC. Cada punto en la curva ROC representa un umbral de clasificación diferente y muestra cómo cambian TPR y FPR en función de ese umbral.

5. Evaluar el rendimiento: La forma y la posición de la curva ROC proporcionan información sobre el rendimiento del modelo. Idealmente, queremos que el modelo tenga un alto TPR y un bajo FPR, lo que se traduce en una curva ROC que se acerque al rincón superior izquierdo del gráfico. Un modelo perfecto tendría un TPR de 1 y un FPR de 0, lo que se representaría por un punto en el extremo superior izquierdo de la curva ROC.

6. Calcular el área bajo la curva (AUC): El área bajo la curva ROC (AUC) es una métrica utilizada para resumir el rendimiento del modelo en un solo valor. Cuanto mayor sea el AUC, mejor será el rendimiento del modelo en términos de capacidad de discriminación entre las clases. Un AUC de 1 indica un modelo perfecto, mientras que un AUC de 0.5 indica un modelo que no es mejor que una predicción aleatoria.

En resumen, la curva ROC es una herramienta para visualizar y evaluar el rendimiento de modelos de clasificación binaria. Permite analizar cómo cambian las tasas de verdaderos positivos y falsos positivos a medida que se varía el umbral de clasificación. La curva ROC y el área bajo la curva (AUC) proporcionan información importante sobre la capacidad de discriminación y el rendimiento general del modelo.

Para la elección entre dos pruebas diagnósticas distintas, se recurre a las curvas ROC, ya que es una medida global e independiente del punto de corte. Por esto, en el ámbito sanitario, las curvas ROC también se denominan curvas de rendimiento diagnóstico.

La elección se realiza mediante la comparación del área bajo la curva (AUC) de ambas pruebas. Esta área posee un valor comprendido entre 0,5 y 1, donde 1 representa un valor diagnóstico perfecto y 0,5 es una prueba sin capacidad discriminatoria diagnóstica. Es decir, si AUC para una prueba diagnóstica es 0,8 significa que existe un 80% de probabilidad de que el diagnóstico realizado a un enfermo sea más correcto que el de una persona sana escogida al azar. Por esto, siempre se elige la prueba diagnóstica que presente un mayor área bajo la curva.

A modo de guía para interpretar las curvas ROC se han establecido los siguientes intervalos para los valores de AUC:

[0.5]: Es como lanzar una moneda.

[0.5, 0.6): Test malo.

[0.6, 0.75): Test regular.

[0.75, 0.9): Test bueno.

[0.9, 0.97): Test muy bueno.

[0.97, 1): Test excelente.

Si la significancia estadística no es una guía útil, ¿qué magnitud de kappa refleja una acuerdo adecuado? Las pautas serían útiles, pero factores distintos al acuerdo pueden influir en su magnitud, lo que dificulta la interpretación de una magnitud dada. Como señalaron Sim y Wright, dos factores importantes son la prevalencia (¿son los códigos equiprobables o varían sus probabilidades) y el sesgo (¿las probabilidades marginales para los dos observadores son similares o diferentes). Otras cosas siendo iguales, los kappas son más altos cuando los códigos son equiprobables. Por otro lado, los kappas son más altos cuando los códigos se distribuyen asimétricamente entre los dos observadores. A diferencia de las variaciones de probabilidad, el efecto del sesgo es mayor cuando Kappa es pequeño que cuando es grande.

Otro factor es el número de códigos. A medida que aumenta el número de códigos, los kappas son más altos. Basado en un estudio de simulación, Bakeman y sus colegas concluyeron que para observadores fallibles, los valores de kappa eran más bajos cuando los códigos eran menos. Y, de acuerdo con la afirmación de Sim y Wright sobre la prevalencia, los kappas eran más altos cuando los códigos eran aproximadamente equiprobables. Así, Bakeman et al. concluyeron que "ningún valor de kappa puede considerarse universalmente aceptable". También proporcionan un programa de computadora que permite a los usuarios calcular valores de kappa especificando el número de códigos, su probabilidad y la precisión del observador. Por ejemplo, dados códigos equiprobables y observadores con una precisión del 85%, los valores de kappa son 0.49, 0.60, 0.66 y 0.69 cuando el número de códigos es 2, 3, 5 y 10, respectivamente.

No obstante, han aparecido pautas de magnitud en la literatura. Quizás la primera fue la de Landis y Koch, quienes caracterizaron los valores < 0 como indicativos de falta de acuerdo, 0-0.20 como ligero, 0.21-0.40 como justo, 0.41-0.60 como moderado, 0.61-0.80 como sustancial y 0.81-1 como acuerdo casi perfecto. Sin embargo, este conjunto de pautas de ninguna manera es universalmente aceptado; Landis y Koch no proporcionaron evidencia para respaldarlo, sino que se basaron en su opinión personal. Se ha observado que estas pautas pueden ser más perjudiciales que útiles. Las pautas igualmente arbitrarias de Fleiss caracterizan los kappas superiores a 0.75 como excelentes, de 0.40 a 0.75 como aceptables a buenos y por debajo de 0.40 como pobres.
