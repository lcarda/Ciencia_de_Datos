K-means

En el algoritmo K-means, la medida de valor de inercia (tambi√©n conocida como suma de cuadrados dentro de grupos, en ingl√©s "within-cluster sum of squares") se utiliza como m√©trica para evaluar la calidad de una partici√≥n de datos.

La f√≥rmula del valor de inercia se calcula sumando las distancias al cuadrado entre cada punto y el centroide de su respectivo grupo. La f√≥rmula general es la siguiente:

Inercia = Œ£[d(x, c)^2]

Donde:

d(x, c) es la distancia entre un punto de datos x y el centroide c de su grupo.
Œ£ representa la suma a lo largo de todos los puntos de datos en el conjunto.
El objetivo del algoritmo K-means es minimizar la inercia, lo que implica que los puntos dentro de cada grupo deben estar cerca de su respectivo centroide. Cuanto menor sea la inercia, m√°s compactos y bien definidos ser√°n los grupos.

El algoritmo K-means busca iterativamente minimizar la inercia, asignando inicialmente centroides aleatorios a los grupos y actualiz√°ndolos en cada iteraci√≥n. Este proceso se repite hasta que los centroides ya no cambien significativamente o se alcance un n√∫mero m√°ximo de iteraciones.

La inercia se utiliza a menudo para seleccionar el n√∫mero √≥ptimo de grupos en K-means. Se puede trazar un gr√°fico de la inercia en funci√≥n del n√∫mero de grupos y buscar un "codo" en el gr√°fico, donde la disminuci√≥n en la inercia se vuelve menos significativa. Este punto puede indicar un equilibrio entre la cantidad de grupos y la calidad de la partici√≥n.

En resumen, la inercia en K-means es una medida de la suma de las distancias al cuadrado entre los puntos de datos y los centroides de sus respectivos grupos. Se utiliza para evaluar la calidad de la partici√≥n y optimizar el n√∫mero de grupos.

La inercia se considera una medida de cu√°n coherentes los clusters son,

heur√≠sticamente que el mejor valor se da cuando se desacelera la reducci√≥n
de la inercia.

#### **Rand index**


La funci√≥n rand_score en scikit-learn es una m√©trica utilizada para evaluar la similitud entre dos agrupamientos (clustering) teniendo en cuenta las etiquetas verdaderas de las muestras. Calcula una medida de similitud considerando todas las parejas de muestras y contando las parejas que se asignan en los mismos o diferentes clusters en los agrupamientos predichos y verdaderos.

El √≠ndice de Rand (Rand index) proporciona una medida de acuerdo entre dos agrupamientos. Un valor cercano a 0 indica una similitud aleatoria, lo que significa que los agrupamientos no est√°n relacionados o no tienen una similitud significativa.
En resumen, el rand_score es una medida de similitud que eval√∫a la calidad de los agrupamientos en funci√≥n de las etiquetas verdaderas. Es √∫til para comparar y evaluar diferentes algoritmos o m√©todos de clustering.

#### **V-measure**

La funci√≥n v_measure_score en scikit-learn es una m√©trica utilizada para evaluar la calidad de un agrupamiento (clustering) en relaci√≥n a las etiquetas verdaderas de las muestras. En resumen, el v_measure_score es una medida que combina la homogeneidad y la completitud para evaluar la calidad global de un agrupamiento. #### **silhouette coefficient** üë§


K-means: An√°lisis de siluetas
- \#\# Para cada punto $i \in C_k$, se crean los indices $a(i)$ de similaridad promedio y $b(i)$ de disimilaridad minima promedio
$$
\begin{gathered}
a(i)=\frac{1}{\left|C_k\right|-1} \sum_{j \in C_k, i \neq j} d(i, j) \quad b(i)=\min _{k \neq l} \frac{1}{\left|C_l\right|} \sum_{j \in C_l} d(i, j) \\
s(i)= \begin{cases}1-a(i) / b(i), & \text { if } a(i)<b(i) \\
0, & \text { if } a(i)=b(i) \\
b(i) / a(i)-1, & \text { if } a(i)>b(i)\end{cases}
\end{gathered}
$$

* El valor 0 indica que la muestra est√° muy cerca del borde de decisi√≥n entre los

El coeficiente de silueta (silhouette coefficient) es una m√©trica utilizada para evaluar la calidad de los agrupamientos (clustering) en funci√≥n de la coherencia interna de los grupos y la separaci√≥n entre ellos. Proporciona una medida de cu√°n bien est√°n definidos los clusters y cu√°n separados est√°n entre s√≠.

El coeficiente de silueta se calcula para cada muestra en el conjunto de datos y var√≠a entre -1 y 1. Los valores m√°s altos indican que una muestra est√° bien asignada a su propio cluster y est√° separada de los otros clusters, mientras que los valores m√°s bajos indican que una muestra puede estar m√°s cerca de los puntos de otros clusters.

La f√≥rmula para calcular el coeficiente de silueta de una muestra combina dos medidas:

Coherencia interna (a): Mide la similitud promedio entre una muestra y las dem√°s muestras dentro del mismo cluster. Cuanto m√°s cercanas est√°n las muestras dentro del cluster, mayor ser√° el valor de coherencia interna.

Separaci√≥n externa (b): Mide la distancia promedio entre una muestra y las muestras de los clusters vecinos m√°s cercanos. Cuanto m√°s lejos est√©n las muestras de otros clusters, mayor ser√° el valor de separaci√≥n externa.

El coeficiente de silueta se calcula como (b - a) / max(a, b) y puede interpretarse de la siguiente manera:

Un valor cercano a 1 indica que la muestra est√° bien asignada a su cluster y est√° separada de los otros clusters.
Un valor cercano a 0 indica que la muestra est√° cerca del l√≠mite de decisi√≥n entre dos clusters.
Un valor cercano a -1 indica que la muestra puede haber sido asignada incorrectamente a su cluster y est√° m√°s cercana a los puntos de otros clusters.
En resumen, el coeficiente de silueta es una m√©trica que eval√∫a la calidad de los agrupamientos en funci√≥n de la coherencia interna y la separaci√≥n entre los clusters. Es √∫til para determinar cu√°n bien est√°n definidos los clusters y si hay una buena separaci√≥n entre ellos.

````python
import numpy as np
from sklearn.datasets import load_digits

data, labels = load_digits(return_X_y=True)
(n_samples, n_features), n_digits = data.shape, np.unique(labels).size


digits = load_digits()

print(f"# digits: {n_digits}; # samples: {n_samples}; # features {n_features}")
````

````python
from time import time
from sklearn import metrics
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler


def bench_k_means(kmeans, name, data, labels):
    """Benchmark to evaluate the KMeans initialization methods.

    Parameters
    ----------
    kmeans : KMeans instance
        A :class:`~sklearn.cluster.KMeans` instance with the initialization
        already set.
    name : str
        Name given to the strategy. It will be used to show the results in a
        table.
    data : ndarray of shape (n_samples, n_features)
        The data to cluster.
    labels : ndarray of shape (n_samples,)
        The labels used to compute the clustering metrics which requires some
        supervision.
    """
    t0 = time()
    estimator = make_pipeline(StandardScaler(), kmeans).fit(data)
    fit_time = time() - t0
    results = [name, fit_time, estimator[-1].inertia_]

    # Define the metrics which require only the true labels and estimator
    # labels
    clustering_metrics = [
        metrics.homogeneity_score,
        metrics.completeness_score,
        metrics.v_measure_score,
        metrics.adjusted_rand_score,
        metrics.adjusted_mutual_info_score,
    ]
    results += [m(labels, estimator[-1].labels_) for m in clustering_metrics]

    # The silhouette score requires the full dataset
    results += [
        metrics.silhouette_score(
            data,
            estimator[-1].labels_,
            metric="euclidean",
            sample_size=300,
        )
    ]

    # Show the results
    formatter_result = (
        "{:9s}\t{:.3f}s\t{:.0f}\t{:.3f}\t{:.3f}\t{:.3f}\t{:.3f}\t{:.3f}\t{:.3f}"
    )
    print(formatter_result.format(*results))
````

````python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

print(82 * "_")
print("init\t\ttime\tinertia\thomo\tcompl\tv-meas\tARI\tAMI\tsilhouette")

kmeans = KMeans(init="k-means++", n_clusters=n_digits, n_init=4, random_state=0)
bench_k_means(kmeans=kmeans, name="k-means++", data=data, labels=labels)

kmeans = KMeans(init="random", n_clusters=n_digits, n_init=4, random_state=0)
bench_k_means(kmeans=kmeans, name="random", data=data, labels=labels)

pca = PCA(n_components=n_digits).fit(data)
kmeans = KMeans(init=pca.components_, n_clusters=n_digits, n_init=1)
bench_k_means(kmeans=kmeans, name="PCA-based", data=data, labels=labels)

print(82 * "_")
````

````python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

# Generating the sample data from make_blobs
# This particular setting has one distinct cluster and 3 clusters placed close
# together.
X, y = make_blobs(
    n_samples=500,
    n_features=2,
    centers=4,
    cluster_std=1,
    center_box=(-10.0, 10.0),
    shuffle=True,
    random_state=1,
)  # For reproducibility

range_n_clusters = [2, 3, 4, 5, 6]

for n_clusters in range_n_clusters:
    # Create a subplot with 1 row and 2 columns
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    # The 1st subplot is the silhouette plot
    # The silhouette coefficient can range from -1, 1 but in this example all
    # lie within [-0.1, 1]
    ax1.set_xlim([-0.1, 1])
    # The (n_clusters+1)*10 is for inserting blank space between silhouette
    # plots of individual clusters, to demarcate them clearly.
    ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])

    # Initialize the clusterer with n_clusters value and a random generator
    # seed of 10 for reproducibility.
    clusterer = KMeans(n_clusters=n_clusters, n_init="auto", random_state=10)
    cluster_labels = clusterer.fit_predict(X)

    # The silhouette_score gives the average value for all the samples.
    # This gives a perspective into the density and separation of the formed
    # clusters
    silhouette_avg = silhouette_score(X, cluster_labels)
    print(
        "For n_clusters =",
        n_clusters,
        "The average silhouette_score is :",
        silhouette_avg,
    )

    # Compute the silhouette scores for each sample
    sample_silhouette_values = silhouette_samples(X, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(
            np.arange(y_lower, y_upper),
            0,
            ith_cluster_silhouette_values,
            facecolor=color,
            edgecolor=color,
            alpha=0.7,
        )

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # The vertical line for average silhouette score of all the values
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # 2nd Plot showing the actual clusters formed
    colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
    ax2.scatter(
        X[:, 0], X[:, 1], marker=".", s=30, lw=0, alpha=0.7, c=colors, edgecolor="k"
    )

    # Labeling the clusters
    centers = clusterer.cluster_centers_
    # Draw white circles at cluster centers
    ax2.scatter(
        centers[:, 0],
        centers[:, 1],
        marker="o",
        c="white",
        alpha=1,
        s=200,
        edgecolor="k",
    )

    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker="$%d$" % i, alpha=1, s=50, edgecolor="k")

    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    plt.suptitle(
        "Silhouette analysis for KMeans clustering on sample data with n_clusters = %d"
        % n_clusters,
        fontsize=14,
        fontweight="bold",
    )

plt.show()
````

#### problema 4

La funci√≥n `AgglomerativeClustering()` de scikit-learn proporciona diferentes opciones de enlace (linkage) y m√©tricas de distancia para realizar el agrupamiento jer√°rquico aglomerativo. A continuaci√≥n, se explican brevemente las nociones de enlace disponibles y las m√©tricas de distancia correspondientes:

1. Enlace completo (complete linkage): Utiliza la distancia m√°xima entre los puntos de dos clusters para determinar la proximidad entre ellos. Es decir, considera la distancia m√°s lejana entre los puntos de dos clusters como la distancia entre los clusters.

2. Enlace promedio (average linkage): Calcula la distancia promedio entre los puntos de dos clusters para determinar la proximidad entre ellos. Toma en cuenta todas las distancias entre los puntos de los dos clusters y las promedia.

3. Enlace simple (single linkage): Utiliza la distancia m√≠nima entre los puntos de dos clusters para determinar la proximidad entre ellos. Es decir, considera la distancia m√°s cercana entre los puntos de dos clusters como la distancia entre los clusters.

Estas opciones de enlace se pueden especificar mediante el par√°metro `linkage` en la funci√≥n `AgglomerativeClustering()`.

En cuanto a las m√©tricas de distancia disponibles, la opci√≥n de enlace en `AgglomerativeClustering()` se combina con diferentes m√©tricas de distancia para calcular las distancias entre los puntos. Algunas de las m√©tricas de distancia comunes son:

- Distancia euclidiana (euclidean): Utiliza la distancia euclidiana entre dos puntos en el espacio.

- Distancia de Manhattan (manhattan): Calcula la distancia entre dos puntos sumando las diferencias absolutas entre sus coordenadas.

- Distancia de Minkowski (minkowski): Generaliza tanto la distancia euclidiana como la de Manhattan. Se puede ajustar el par√°metro `p` para controlar la forma de la m√©trica. Cuando `p=2`, es equivalente a la distancia euclidiana, y cuando `p=1`, es equivalente a la distancia de Manhattan.

Estas m√©tricas de distancia se especifican mediante el par√°metro `affinity` en la funci√≥n `AgglomerativeClustering()`.

En resumen, al utilizar la funci√≥n `AgglomerativeClustering()`, puedes elegir diferentes opciones de enlace (complete, average, single) y combinarlas con diversas m√©tricas de distancia (euclidean, manhattan, minkowski) para realizar el agrupamiento jer√°rquico aglomerativo seg√∫n tus necesidades y caracter√≠sticas de los datos.

lo que aparese en sklearn


linkage{‚Äòward‚Äô, ‚Äòcomplete‚Äô, ‚Äòaverage‚Äô, ‚Äòsingle‚Äô}, default=‚Äôward‚Äô
Este par√°metro determina el criterio de enlace a utilizar en el algoritmo de agrupamiento jer√°rquico aglomerativo. El criterio de enlace determina qu√© distancia se utilizar√° entre conjuntos de observaciones. El algoritmo fusionar√° los pares de clusters que minimicen este criterio.

- ‚Äòward‚Äô minimiza la varianza de los clusters que se fusionan.
- ‚Äòaverage‚Äô utiliza el promedio de las distancias de cada observaci√≥n de los dos conjuntos.
- ‚Äòcomplete‚Äô o ‚Äòmaximum‚Äô linkage utiliza las distancias m√°ximas entre todas las observaciones de los dos conjuntos.
- ‚Äòsingle‚Äô utiliza el m√≠nimo de las distancias entre todas las observaciones de los dos conjuntos.

En resumen, el par√°metro "linkage" determina qu√© medida de distancia se utilizar√° para fusionar los clusters y minimizar el criterio correspondiente.

````python
# Authors: Gael Varoquaux
# License: BSD 3 clause (C) INRIA 2014

from time import time

import numpy as np
from matplotlib import pyplot as plt

from sklearn import manifold, datasets

digits = datasets.load_digits()
X, y = digits.data, digits.target
n_samples, n_features = X.shape

np.random.seed(0)


# ----------------------------------------------------------------------
# Visualize the clustering
def plot_clustering(X_red, labels, title=None):
    x_min, x_max = np.min(X_red, axis=0), np.max(X_red, axis=0)
    X_red = (X_red - x_min) / (x_max - x_min)

    plt.figure(figsize=(6, 4))
    for digit in digits.target_names:
        plt.scatter(
            *X_red[y == digit].T,
            marker=f"${digit}$",
            s=50,
            c=plt.cm.nipy_spectral(labels[y == digit] / 10),
            alpha=0.5,
        )

    plt.xticks([])
    plt.yticks([])
    if title is not None:
        plt.title(title, size=17)
    plt.axis("off")
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])


# ----------------------------------------------------------------------
# 2D embedding of the digits dataset
print("Computing embedding")
X_red = manifold.SpectralEmbedding(n_components=2).fit_transform(X)
print("Done.")

from sklearn.cluster import AgglomerativeClustering

for linkage in ("ward", "average", "complete", "single"):
    clustering = AgglomerativeClustering(linkage=linkage, n_clusters=10)
    t0 = time()
    clustering.fit(X_red)
    print("%s :\t%.2fs" % (linkage, time() - t0))

    plot_clustering(X_red, clustering.labels_, "%s linkage" % linkage)


plt.show()
````

````python
import numpy as np

from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering


def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)


iris = load_iris()
X = iris.data

# setting distance_threshold=0 ensures we compute the full tree.
model = AgglomerativeClustering(distance_threshold=0, n_clusters=None)

model = model.fit(X)
plt.title("Hierarchical Clustering Dendrogram")
# plot the top three levels of the dendrogram
plot_dendrogram(model, truncate_mode="level", p=3)
plt.xlabel("Number of points in node (or index of point if no parenthesis).")
plt.show()
````

AgglomerativeClustering() es un algoritmo de agrupamiento jer√°rquico aglomerativo que se utiliza para agrupar un conjunto de datos en clusters. El algoritmo comienza considerando cada punto de datos como un cluster individual y luego fusiona gradualmente los clusters vecinos en funci√≥n de una medida de distancia o similitud.

El algoritmo sigue el enfoque "bottom-up" (de abajo hacia arriba) en el cual en cada paso se fusionan los dos clusters m√°s cercanos seg√∫n el criterio de enlace especificado. Esto se repite hasta que todos los puntos de datos est√©n agrupados en un solo cluster o se cumpla un criterio de parada predefinido.

La funci√≥n AgglomerativeClustering() en la biblioteca scikit-learn implementa este algoritmo. Recibe varios par√°metros, como el n√∫mero de clusters a buscar, el criterio de enlace a utilizar, la m√©trica de distancia, etc. Una vez que se realiza el ajuste del modelo, se pueden obtener las etiquetas de cluster asignadas a cada punto de datos y utilizarlas para an√°lisis posteriores o visualizaci√≥n.

En resumen, AgglomerativeClustering() es una funci√≥n que realiza el agrupamiento jer√°rquico aglomerativo en un conjunto de datos, fusionando gradualmente los clusters seg√∫n un criterio de enlace y produciendo las asignaciones de cluster correspondientes.
